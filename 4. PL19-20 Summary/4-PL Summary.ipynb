```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

fixg,ax = plt.subplots(figsize=(10,8))
p1 = plt.bar(ind, menMeans, width, yerr=menStd,color='r')
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd,color='b')

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

```python
std_stats = pd.read_csv('D:\Football\Data\FBref\PL\Standard-stats.csv',header=[0,1])
std_stats.head()
```

```python
std_stats.columns = [f'{i}{j}' for i, j in std_stats.columns]
```

```python
std_stats.columns
```

```python
yellow_card_data = std_stats.sort_values(by='PerformanceCrdY')[[' Team','PerformanceCrdY','PerformanceCrdR']]
```

```python
penalty_data = std_stats.sort_values(by=['PerformancePK','PerformancePKatt'],ascending=False)[[' Team','PerformancePK','PerformancePKatt']]
```

```python
penalty_data['PKMissed'] = penalty_data['PerformancePKatt'] - penalty_data['PerformancePK']
penalty_data
```

```python
csfont = {'fontname':'serif'}

fig,ax = plt.subplots(figsize=(8,9))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# ax.spines['bottom'].set_visible(False)

p1 = plt.barh(np.arange(20),penalty_data['PerformancePK'],height=0.7,color="#00aa00",edgecolor="#000000")
p2 = plt.barh(np.arange(20),penalty_data['PKMissed'],left=penalty_data['PerformancePK'],height=0.7,color="#ff0000",edgecolor="#000000")


plt.xlabel('Goals',fontsize=16,labelpad=10,**csfont)
plt.title('Penalty Kicks',fontsize=22,**csfont)

plt.xticks(np.arange(11),**csfont,fontsize=14)
plt.yticks(np.arange(20), penalty_data[' Team'],**csfont,fontsize=14)

plt.legend((p1[0], p2[0]), ('Scored', 'Missed'),fontsize=15)
# plt.grid(linestyle='-',color='white',alpha=0.5)

plt.annotate('Viz by @gkgokul10', xy=(0.7,-0.08), xycoords='axes fraction',fontsize='14')

plt.savefig('D:/Pen-kics.png',dpi=1000,bbox_inches = "tight")
plt.show()
```

```python
SCA_GCA_stats= pd.read_csv('D:\Football\Data\FBref\PL\SCA-GCA.csv',header=[0,1])
SCA_GCA_stats.head()    
```

```python
SCA_GCA_stats.columns = [f'{i}{j}' for i, j in SCA_GCA_stats.columns]
SCA_GCA_stats.columns
```

```python
SCA_GCA = SCA_GCA_stats[[' Team','SCASCA90','GCAGCA90']]
SCA_GCA
```

```python
np.median(SCA_GCA['SCASCA90'])
```

```python
csfont = {'fontname':'serif'}

fig,ax = plt.subplots(figsize=(12,12))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.yticks(np.arange(0,4.5,0.25))
plt.xticks(np.arange(10,34,1))

plt.scatter(SCA_GCA['SCASCA90'],SCA_GCA['GCAGCA90'],color='red')

for ind,data in SCA_GCA.iterrows():
    if(data[' Team']=='Bournemouth' or data[' Team']=='Norwich City' or data[' Team']=='Watford'or data[' Team']=='West Ham'or data[' Team']=='Everton'):
        plt.annotate(data[' Team'], xy=(data['SCASCA90']-0.8,data['GCAGCA90']-0.07), xycoords='data',fontsize=12,**csfont)
    else:
        plt.annotate(data[' Team'], xy=(data['SCASCA90']-1,data['GCAGCA90']+0.03), xycoords='data',fontsize=12,**csfont)

        
        
plt.xlabel('Shot-Creating Actions per 90',fontsize=16, labelpad=10,**csfont)
plt.ylabel('Goal-Creating Actions per 90',fontsize=16, labelpad=10,**csfont)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.title('Attacking Actions per 90',fontsize=22,**csfont)
plt.grid(linestyle='-',color='white',alpha=0.25)

plt.annotate('Viz by @gkgokul10', xy=(0.8,-0.07), xycoords='axes fraction',fontsize='14')

plt.savefig('D:/Attacking-Actions-per90.png',dpi=700,bbox_inches = "tight")
plt.show()
```

```python
Sqd_def_stats= pd.read_csv('D:\Football\Data\FBref\PL\Squad-def-actions.csv',header=[0,1])
Sqd_def_stats.head()    
```

```python
Sqd_def_stats.columns = [f'{i}{j}' for i, j in Sqd_def_stats.columns]
Sqd_def_stats.columns
```

```python
att3_press = Sqd_def_stats[[' Team','PressuresPress','Pressures%','PressuresDef 3rd', 'PressuresMid 3rd',
       'PressuresAtt 3rd']]

# df['Discounted_Price'] = df.apply(lambda row: row.Cost - 
#                                   (row.Cost * 0.1), axis = 1) 

# att3_press['%inAtt3rd'] = att3_press.apply(lambda row: row['PressuresAtt 3rd']/row['PressuresPress'], axis=1) 

att3_press['%inDef3rd'] = att3_press['PressuresDef 3rd']*100/att3_press['PressuresPress']
att3_press['%inMid3rd'] = att3_press['PressuresMid 3rd']*100/att3_press['PressuresPress']
att3_press['%inAtt3rd'] = att3_press['PressuresAtt 3rd']*100/att3_press['PressuresPress']

att3_press['%inAtt3rd'] = att3_press['%inAtt3rd'].astype(float).round(1)
att3_press['%inMid3rd'] = att3_press['%inMid3rd'].astype(float).round(1)
att3_press['%inDef3rd'] = att3_press['%inDef3rd'].astype(float).round(1)

att3_press.sort_values(by='%inAtt3rd')
att3_press
```

```python
press_stats = att3_press.sort_values(by='PressuresPress')
press_stats
```

```python
import matplotlib.colors as mcolors

csfont = {'fontname':'serif'}

fig,ax = plt.subplots(figsize=(14,7))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

clist = [(0, "#cccc00"),(1, "#ffff00")]
rvb1 = mcolors.LinearSegmentedColormap.from_list("", clist)

plt.xlabel('No. of Pressing action',fontsize=16,labelpad=10,**csfont)

plt.xticks(np.arange(0,6000,250),fontsize=10)
plt.yticks(np.arange(20), press_stats[' Team'],**csfont,fontsize=14)

p1 = plt.barh(np.arange(20),press_stats['PressuresAtt 3rd'],height=0.7,color='#cccc00',edgecolor="#000000")
p2 = plt.barh(np.arange(20),press_stats['PressuresMid 3rd'],left=press_stats['PressuresAtt 3rd'],height=0.7,color='#ff8800',edgecolor="#000000",alpha=0.9)
p3 = plt.barh(np.arange(20),press_stats['PressuresDef 3rd'],left=press_stats['PressuresAtt 3rd']+press_stats['PressuresMid 3rd'],height=0.7,color='red',edgecolor="#000000",alpha=0.9)

plt.title('Pressing Actions',fontsize=22,**csfont)
plt.legend((p1[len(p1)-1], p2[len(p2)-1],p3[len(p3)-1]), ('Att 3rd', 'Mid 3rd','Def 3rd'),fontsize=15)
plt.grid(linestyle='-',color='white',alpha=0.3)

plt.annotate('Viz by @gkgokul10', xy=(0.8,-0.1), xycoords='axes fraction',fontsize='14')


plt.savefig('D:/Press%OVR.png',dpi=800,bbox_inches = "tight")
plt.show()
```
